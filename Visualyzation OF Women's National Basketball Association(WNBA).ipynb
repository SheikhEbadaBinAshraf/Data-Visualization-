```python
import pandas as pd
import numpy as sns
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
wnba = pd.read_csv("wnba.csv")
```

```python
wnba["Experience"].value_counts(dropna = False)
```

#### R stands for `Rukee` Or `Dabue`

```python
# function help for getting insite 

def category(x):
    if pd.isnull(x):
        return "Rookie"
    elif 0 < x <= 3:
        return "Little Experience"
    elif 3 < x <= 5:
        return "Experienced"
    elif 5 < x <= 10:
        return "Very Experienced"
    elif x > 10:
        return "Veteran"
```

```python
# Applying above function

wnba["Exp_ordinal"] =  pd.to_numeric(wnba["Experience"], errors = "coerce").apply(category)
```

```python
# Now we can get insite

wnba["Exp_ordinal"].value_counts()
```

```python
# Show in squence

value = wnba["Exp_ordinal"].value_counts().iloc[[3,0,2,1,4]]
```

```python
value
```

```python
# Visualizaion by Pandas
value.plot(kind = "bar", rot = 45, colormap = plt.cm.PRGn, title = "Number of Players in WNBA by Level of Experience")

plt.show()
```

```python
# Pie Chart with percentage

value.plot.pie(colormap = plt.cm.viridis, title = "Percentage of Players in WNBA by Level of Experience",
              startangle = 65, autopct ="%.1f%%")
plt.ylabel("")

plt.show()
```

# Summary: `WNBA Player Analysis - Experience and Position`

### This WNBA data visualization project centers on a detailed exploration of player demographics, experience levels, and playing positions. The visualizations unearthed compelling insights into the diverse landscape of player composition within the league:

## 1. Experience Level Breakdown:
### Visual representations categorize players into distinct experience levels—Rookie, Little Experience, Experienced, Very-Experienced, and Veteran. The project delves into the distribution of players across these categories, offering a nuanced understanding of the league's player composition.

## 2. Positional Analysis:
### The project examined the distribution of players across different positions—Forwards (F), Centers (C), Forwards/Centers (F/C), Guards (G), and Forwards/Guards (F/G). Insights into the prevalence of each position shed light on the strategic diversity employed by teams and the versatility of players.
